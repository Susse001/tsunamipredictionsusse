```python
import pandas as pd
import numpy as np

# Load datasets
eq = pd.read_csv('earthquake_data_tsunami.csv')
ts = pd.read_csv('tsunamis.csv')

def tsunami_occurred(eq_year, eq_month, eq_mag, mag_tolerance=0.1):
    subset = ts[(ts['Year'] == eq_year) & (ts['Mo'] == eq_month)]
    if subset.empty:
        return False
    mags = subset['EarthquakeNanMagnitude'].dropna()
    return np.any(np.abs(mags - eq_mag) <= mag_tolerance)

eq['tsunami'] = eq.apply(
    lambda r: 1 if tsunami_occurred(r["Year"], r["Month"], r["magnitude"]) else 0,
    axis=1
)

eq.head()
eq.to_csv("earthquake_data_tsunami2.csv", index=False)
```
